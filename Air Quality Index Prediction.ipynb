<h1>Project: Air Quality India Project</h1>

<h3>Probelm Statement</h3>
Major Indian cities face a "Severe" air crisis with PM2.5 levels exceeding WHO limits by 15x during winter peaks. We aim to build a predictive model that achieves a MAE of < 15 AQI points, providing a 24-hour early warning system to reduce population exposure by 20% through proactive policy action.

<h3>Goal & Aim</h3>
Project Goal: To build a high-precision regression framework capable of predicting daily city-level AQI with an MAE < 15 points using historical CPCB pollutant data.

Project Aim: To provide municipalities with a 24-hour predictive early-warning window, shifting public health responses from reactive damage control to proactive exposure mitigation.

```python
### Importing the libraries and loading the data


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./Air Quality India Project/air_quality_cleaned.csv")





```

```python
### The target is Numberical. To show if the target is balanced or not. plotting the chart below
## I used AI to generate the plot.

plt.figure(figsize=(10, 6))

# 3. Plot the histogram distribution
plt.hist(df['AQI'], bins=40, color='skyblue', edgecolor='black', alpha=0.7)

# 4. Add statistical indicator lines (Mean and Median)
mean_aqi = df['AQI'].mean()
median_aqi = df['AQI'].median()

plt.axvline(mean_aqi, color='red', linestyle='dashed', linewidth=1.5, 
            label=f"Mean: {mean_aqi:.0f}")
plt.axvline(median_aqi, color='green', linestyle='dashed', linewidth=1.5, 
            label=f"Median: {median_aqi:.0f}")

# 5. Add labels and title
plt.title('Distribution of Air Quality Index (Target Variable)', fontsize=14)
plt.xlabel('AQI', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()

# 6. Show the final layout cleanly
plt.tight_layout()
plt.show()
```

```python
## exploring the outliers
## We found outliers in each feature. We will have to deal with them after deeper understanding.

# 1. Filter out only numerical features
numerical_cols = df.select_dtypes(include='number').columns

# 2. Set up the matplotlib figure grid
num_features = len(numerical_cols)
fig, axes = plt.subplots(nrows=(num_features + 2) // 3, ncols=3, figsize=(15, 4 * ((num_features + 2) // 3)))
axes = axes.flatten()  # Flatten to easily iterate with a single index

# 3. Loop through and create a boxplot for each feature
for i, col in enumerate(numerical_cols):
    sns.boxplot(x=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)
    axes[i].set_xlabel('')

# 4. Hide any unused subplots in the grid
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

```

```python
### To take a look at the correlation with each other, below plotting numerical values 
## Categorical features need Annova since we study categorical to numberic target.


the_list = df.select_dtypes(include='number').columns
corr_matrix = df[the_list].corr()

# 1. Set a larger figure size (Width, Height in inches)
plt.figure(figsize=(10, 8))

# 2. Plot the heatmap with adjusted font size and formatting
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})

# 3. Prevent labels from getting cut off at the edges
plt.tight_layout()
plt.show()

## It seems most features not too much correlated features with others. Strongest CO, PMI 2.5 with target.
```
