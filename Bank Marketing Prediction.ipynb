```python
### Importing the libraries and loading the data


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./Bank Marketing Project/bank-full.csv", sep =';')
df = df.rename(columns={'y': 'deposit'})




```

```python
# Calculate percentages
percentages = df['deposit'].value_counts(normalize=True) * 100

# Set up the plot
plt.figure(figsize=(6, 5))
ax = sns.barplot(x=percentages.index, y=percentages.values, palette='Set2')

# Add labels and formatting
plt.title('Distribution of Target Variable (deposit)', fontsize=14, fontweight='bold')
plt.xlabel('Subscribed to Deposit?', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.ylim(0, 100) # Keep scale to 100% for context

# Annotate the actual percentages on top of the bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width() / 2., p.get_height() + 1.5), 
                ha='center', va='center', fontsize=11, fontweight='bold', color='black')

plt.tight_layout()
plt.show()
```

```python
### To take a look at the correlation with target, below plotting numerical values 
## Categorical features need Annova since we study categorical to numberic target.


the_list = df.select_dtypes(include='number').columns
corr_matrix = df[the_list].corr()

sns.heatmap(corr_matrix, annot=True)

## It seems no strong correlated features with others. 
```

```python
## exploring the outliers
## We found age, balance, duration, campaign, pdays, previous have some outliers

# 1. Filter out only numerical features
numerical_cols = df.select_dtypes(include='number').columns

# 2. Set up the matplotlib figure grid
num_features = len(numerical_cols)
fig, axes = plt.subplots(nrows=(num_features + 2) // 3, ncols=3, figsize=(15, 4 * ((num_features + 2) // 3)))
axes = axes.flatten()  # Flatten to easily iterate with a single index

# 3. Loop through and create a boxplot for each feature
for i, col in enumerate(numerical_cols):
    sns.boxplot(x=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)
    axes[i].set_xlabel('')

# 4. Hide any unused subplots in the grid
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

```
