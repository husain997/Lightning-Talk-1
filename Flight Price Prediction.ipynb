<h1>Project: Flight Price Prediction</h1>

<h3>Probelm Statement</h3>
Individual travelers and travel agencies face extreme flight price volatility in the Indian aviation market, with ticket prices surging by up to 250% within 48 hours of departure. This leads to an average budgeting discrepancy of 15-20%. We aim to deploy a regression model to reduce price forecasting errors by 25% and achieve a target accuracy of 95% within the next quarter.

<h3>Goal & Aim</h3>
To develop a robust machine learning Regression Model that forecasts ticket prices using features like lead time, class, and duration. Our aim is to provide a decision-support tool that enables users to optimize booking times and save up to 30% on airfare costs.

```python
### Importing the libraries and loading the data


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./Flights Price Prediction Project/flights_data.csv")

#dropping the column Unnamed: 0' as it is index only.
df = df.drop(columns=['Unnamed: 0'])




```

```python
#Exploring the dataset
df.head()
```

```python
#Exploring the types and if any to change to right data type
df.dtypes
```

```python
### Exploring the number of nulls found in the dataset. 

df.isnull().sum()
```

```python
#Taking a look at descriptive statistics and later go deeper into analysis
df.describe()
```

```python
#To check the distribution and get rough understanding if there is outliers.. Exploring all features value counts.
df['days_left'].value_counts
df['duration'].value_counts
```

```python
### The target is Numberical. To show if the target is balanced or not. plotting the chart below
## I used AI to generate the plot.

plt.figure(figsize=(10, 6))

# 3. Plot the histogram distribution
plt.hist(df['price'], bins=40, color='skyblue', edgecolor='black', alpha=0.7)

# 4. Add statistical indicator lines (Mean and Median)
mean_price = df['price'].mean()
median_price = df['price'].median()

plt.axvline(mean_price, color='red', linestyle='dashed', linewidth=1.5, 
            label=f"Mean: {mean_price:.0f} INR")
plt.axvline(median_price, color='green', linestyle='dashed', linewidth=1.5, 
            label=f"Median: {median_price:.0f} INR")

# 5. Add labels and title
plt.title('Distribution of Flight Prices (Target Variable)', fontsize=14)
plt.xlabel('Price (in INR)', fontsize=12)
plt.ylabel('Frequency (Number of Flights)', fontsize=12)
plt.legend()

# 6. Show the final layout cleanly
plt.tight_layout()
plt.show()
```

```python
### To take a look at the correlation with target, below plotting numerical values with the target (price)
## Categorical features need Annova since we study categorical to numberic target.

the_list = df[['price', 'days_left', 'duration']].corr()

sns.heatmap(the_list, annot=True)
plt.show()
```

```python
# Create a side-by-side subplot to look at prices when combining both "class" economy and business. Then another plot to see ho outliers reduced.

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Overall Price Boxplot (Shows massive "outliers")
sns.boxplot(data=df, y='price', ax=axes[0], color='salmon')
axes[0].set_title('Overall Price Boxplot\n(Looks like many outliers)')
axes[0].set_ylabel('Price (INR)')

# Plot 2: Price Boxplot by Class (Explains the outliers!)
sns.boxplot(data=df, x='class', y='price', ax=axes[1], palette='Set2')
axes[1].set_title('Price Boxplot Separated by Class\n(Outliers decreased)')
axes[1].set_ylabel('Price (INR)')
axes[1].set_xlabel('Cabin Class')

plt.tight_layout()
plt.show()
```

```python
##Taking a look at outliers of duration feature.

fig, ax = plt.subplots(figsize=(6, 5))

# Pass 'ax' instead of 'axes[0]'
sns.boxplot(data=df, y='duration', ax=ax, color='salmon')
ax.set_title('Overall Duration Boxplot')
ax.set_ylabel('Duration')

plt.tight_layout()
plt.show()
```

```python
##Taking a look at outliers of lead_days feature.

fig, ax = plt.subplots(figsize=(6, 5))

# Pass 'ax' instead of 'axes[0]'
sns.boxplot(data=df, y='days_left', ax=ax, color='salmon')
ax.set_title('Overall leads in days Boxplot')
ax.set_ylabel('Lead days')

plt.tight_layout()
plt.show()
```
